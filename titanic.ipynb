```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.fillna(df['Age'].mean(), inplace=True)
df['Age'] = df['Age'].astype(int)   
```

```python
df['Age'].isnull().sum()
```

```python
inputs = df.drop(labels=['Survived', 'PassengerId', 'Name','SibSp', 'Parch','Ticket','Cabin', 'Embarked'], axis='columns')
target = df['Survived']
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
```

```python
inputs['Sex_n'] = le_sex.fit_transform(inputs['Sex'])
inputs
```

```python
inputs = inputs.drop(labels=['Age_n'], axis='columns')
inputs
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs, target)
```

```python
model.score(inputs, target)
```

```python
model.predict([[3,47, 7,0],[3,21,6,1]])
```
